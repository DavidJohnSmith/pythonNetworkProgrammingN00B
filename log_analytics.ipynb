```python
import re
import pandas as pd
from collections import Counter

data = open('www-access.log').readlines()
colnames   = ('host','referrer','user','datetime',
              'method','request','proto','status','bytes')
logpats = r'(\S+) (\S+) (\S+) \[(.*?)\] '\
          r'"(\S+) (\S+) (\S+)" (\S+) (\S+)'
logpat = re.compile(logpats)

#########################################
# generator processing
#########################################

groups  = (logpat.match(line) for line in data)
tuples  = (g.groups() for g in groups if g)
log        = (dict(zip(colnames,t)) for t in tuples)

# generator function for ripping out all the host
def host(iterable):
    for item in iterable:
        yield item['host']

#########################################
# generator processing into a function
#########################################

hosts   = host(log)
result = Counter(hosts).most_common()
ip =[]
count = []
for key, value in result:
    ip.append(key)
    count.append(value)
    
ts= pd.Series(count, ip)
ts.plot(kind='bar')
```
